```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
names = []
sectors = []
country = []

#loop 144 times
for count in range(1,145):

    url = "https://www.cosmoprof.com/en/visit/exhibitors-directory/exhibitors-list/?y=0&page={}".format(count)
    page_response = requests.get(url, timeout=5)

    page_content = BeautifulSoup(page_response.content, "html.parser")

    company_container = page_content.find_all("div", class_="container brand")

    for temp in company_container:
        test = temp.find_all("div", class_="row")


        for row in test:
            company = row.find_all("div", class_ = "col-md-3 result-item mb-40")


            for company_info in company:

                company_name=company_info.find('h3')
                names.append(company_name.text)
                #print (company_name.text)

                company_data = company_info.find_all('li')

                country_name = company_data[0].text[9:]
                #print(country_name)




                if (country_name == "Italia"):
                    country.append("Italy")

                elif (country_name == "Francia"):
                    country.append("France")

                elif (country_name == "Regno Unito"):
                    country.append("UK")

                elif (country_name == "Corea"): 
                    country.append("Korea")
                elif (country_name == "Giappone"): 
                    country.append("Japan")
                elif (country_name == "Repubblica Popolare Cinese"):
                    country.append("China")
                elif (country_name == "Stati Uniti d'America") : 
                    country.append("USA")
                else: 
                    country.append(country_name)


                #Italia|Francia|Regno Unito| Corea | Giappone | Repubblica Popolare Cinese | Stati Uniti d'America
                #country.append(country_name)



                sector_name = company_data[1].text[7:]
                #print(sector_name)
                sectors.append(sector_name)


       #end of loop
    print("loop ended!")
            
       
        
```

```python
df = pd.DataFrame({'Company':names, 'Country':country, 'Sector':sectors})
df
```
